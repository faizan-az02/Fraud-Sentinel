```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import LabelEncoder
import os
```

```python
INPUT_PATH = "D:\Github\Fraud-Sentinel\data\processed\merged_data.csv"
OUTPUT_PATH = "data/processed/clean_data.csv"


def load_data():
    print("Loading merged data...")
    df = pd.read_csv(INPUT_PATH)
    print("Shape:", df.shape)
    return df
```

```python
df = load_data()

print(df.head())
```

```python
count = 0
threshold = 0.80
for c in df.columns:
    if df[c].isnull().sum()/len(df) > threshold:
        count += 1
        print(f"{c}: Empty percentage: {df[c].isnull().sum() / len(df) * 100:.2f}%")
print(f"Total columns with more than {threshold*100:.0f}% empty values: {count}")
```

```python
count_dropped = 0
for c in df.columns:
    if df[c].isnull().sum()/len(df) > threshold:
        count_dropped += 1
        df.drop(columns=[c], inplace=True)

print(f"Dropped {count_dropped} columns with more than {threshold*100:.0f}% empty values.")
```

```python
count = 0
threshold = 0.5
for c in df.columns:
    if df[c].isnull().sum()/len(df) > threshold:
        count += 1
        print(f"{c}: Empty percentage: {df[c].isnull().sum() / len(df) * 100:.2f}%")
print(f"Total columns with more than {threshold*100:.0f}% empty values: {count}/{len(df.columns)}")
```

```python
# Fill missing values:
# - Numerical (int/float) → median
# - Categorical (object) → "missing"

for col in df.columns:
    if df[col].dtype in ["int64", "float64"]:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna("missing")

print("Missing values handled!")
```

```python
df = df.drop(["TransactionID", "TransactionDT"], axis=1, errors="ignore")
```
